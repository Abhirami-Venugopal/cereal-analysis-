```python
#calculating the average calories of all cereals made by each brand
rows_list = []
for MFR in df['mfr'].unique():
    calorieSum = df.loc[df['mfr'] == MFR]['calories'].sum()
    numCerealsByBrand = len(df.loc[df['mfr']==MFR])
    avgCalorieByBrand = calorieSum/numCerealsByBrand;
    dict1 = {'mfr': MFR, 'Calorie By Brand': avgCalorieByBrand, 'numCerealsByBrand': numCerealsByBrand};
    rows_list.append(dict1)
```

```python
df1 = pd.DataFrame(rows_list)
df1.sort_values(['Calorie By Brand'], ascending = True, inplace = True)
df1
```

```python
#Based on the chart above, we can see that Nabisco and Quaker Oats 
#have the highest average calories. 
```

```python
plt.figure(figsize=(10,10))
ax = sns.barplot(df1['mfr'], df1['Calorie By Brand'])
ax.set_xticklabels(ax.get_xticklabels(), rotation = 40, ha = "right")
```
